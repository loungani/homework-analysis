```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
homework = pd.read_excel("Homework.xlsx")
```

```python
homework.head()
```

```python
classes = homework.groupby('Class')['~ Hours Invested'].sum()
classes
```

```python
index = classes.index
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(np.arange(len(classes)), classes, 0.35)
ax.set_xticks(np.arange(len(classes)))
ax.set_xticklabels(index)
plt.title("Hours Invested by Class")

labels = classes
for rect, label in zip(bars, labels):
    height = bars.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height, label,
            ha='center', va='bottom')
    
plt.show()
```
